# Praktikum PCD 7 - Filter Rerata

```python
import matplotlib.pyplot as plt  # Mengimport modul matplotlib.pyplot dengan alias plt
%matplotlib inline  # Mengaktifkan mode inline untuk menampilkan plot secara interaktif di notebook

from skimage import data  # Mengimport modul data dari paket skimage
from skimage.io import imread  # Mengimport fungsi imread dari modul io dalam paket skimage
from skimage.color import rgb2gray  # Mengimport fungsi rgb2gray dari modul color dalam paket skimage
import numpy as np  # Mengimport modul numpy dengan alias np
```

Load & Plot Input Image

```python
citra1 = imread(fname="gedung.tif")  # Membaca citra dengan nama file "gedung.tif" menggunakan fungsi imread dan menyimpannya ke variabel citra1
citra2 = imread(fname="mobil.tif")  # Membaca citra dengan nama file "mobil.tif" menggunakan fungsi imread dan menyimpannya ke variabel citra2

print('Shape citra 1 : ', citra1.shape)  # Menampilkan dimensi citra 1 dengan menggunakan atribut shape pada variabel citra1
print('Shape citra 2 : ', citra2.shape)  # Menampilkan dimensi citra 2 dengan menggunakan atribut shape pada variabel citra2

fig, axes = plt.subplots(1, 2, figsize=(10, 10))  # Membuat objek gambar dengan menggunakan subplots untuk menampilkan dua gambar secara horizontal, dengan ukuran gambar 10x10
ax = axes.ravel()  # Melakukan perataan (flattening) pada array objek sumbu

ax[0].imshow(citra1, cmap = 'gray')  # Menampilkan citra 1 pada sumbu ke-0 dengan menggunakan imshow dan menerapkan skema warna "gray"
ax[0].set_title("Citra 1")  # Mengatur judul sumbu ke-0 menjadi "Citra 1"

ax[1].imshow(citra2, cmap = 'gray')  # Menampilkan citra 2 pada sumbu ke-1 dengan menggunakan imshow dan menerapkan skema warna "gray"
ax[1].set_title("Citra 2")  # Mengatur judul sumbu ke-1 menjadi "Citra 2"
```

Menyiapkan variable output

```python
copyCitra1 = citra1.copy().astype(float)  # Membuat salinan citra1 dan mengubah tipe datanya menjadi float dengan menggunakan method copy dan astype(float)
copyCitra2 = citra2.copy().astype(float)  # Membuat salinan citra2 dan mengubah tipe datanya menjadi float dengan menggunakan method copy dan astype(float)

m1, n1 = copyCitra1.shape  # Menyimpan ukuran citra1 (jumlah baris dan kolom) ke dalam variabel m1 dan n1
output1 = np.empty([m1, n1])  # Membuat array kosong dengan ukuran m1 x n1 sebagai output1

m2, n2 = copyCitra2.shape  # Menyimpan ukuran citra2 (jumlah baris dan kolom) ke dalam variabel m2 dan n2
output2 = np.empty([m2, n2])  # Membuat array kosong dengan ukuran m2 x n2 sebagai output2

print('Shape copy citra 1 : ', copyCitra1.shape)  # Menampilkan dimensi copyCitra1 menggunakan atribut shape
print('Shape output citra 1 : ', output1.shape)  # Menampilkan dimensi output1 menggunakan atribut shape

print('m1 : ', m1)  # Menampilkan nilai m1
print('n1 : ', n1)  # Menampilkan nilai n1
print()

print('Shape copy citra 2 : ', copyCitra2.shape)  # Menampilkan dimensi copyCitra2 menggunakan atribut shape
print('Shape output citra 3 : ', output2.shape)  # Menampilkan dimensi output2 menggunakan atribut shape
print('m2 : ', m2)  # Menampilkan nilai m2
print('n2 : ', n2)  # Menampilkan nilai n2
print()
```

Proses Filter Rerata Pada Citra Input 1

```python
for baris in range(0, m1-1):  # Melakukan iterasi pada setiap baris citra1 kecuali baris terakhir.
    for kolom in range(0, n1-1):  # Melakukan iterasi pada setiap kolom citra1 kecuali kolom terakhir.
        a1 = baris  # Menginisialisasi variabel a1 dengan nilai baris saat ini.
        b1 = kolom  # Menginisialisasi variabel b1 dengan nilai kolom saat ini.
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]  # Menghitung jumlah piksel di sekitar piksel saat ini.
        output1[a1, b1] = (1/9 * jumlah)  # Menghitung rata-rata dari jumlah piksel dan menetapkan nilainya pada output1[a1, b1].
```

Proses Filter Rerata Pada Citra Input 2

```python
for baris1 in range(0, m2-1):  # Melakukan iterasi pada setiap baris citra2 kecuali baris terakhir.
    for kolom1 in range(0, n2-1):  # Melakukan iterasi pada setiap kolom citra2 kecuali kolom terakhir.
        a1 = baris1  # Menginisialisasi variabel a1 dengan nilai baris1 saat ini.
        b1 = kolom1  # Menginisialisasi variabel b1 dengan nilai kolom1 saat ini.
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1-1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1]  # Menghitung jumlah piksel di sekitar piksel saat ini.
        output2[a1, b1] = (1/9 * jumlah)  # Menghitung rata-rata dari jumlah piksel dan menetapkan nilainya pada output2[a1, b1].
```

Plot Citra Input dan Output Hasil dari Filter Rerata

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # Membuat objek subplot dengan ukuran 2x2 dan ukuran gambar 10x10.
ax = axes.ravel()  # Meratakan objek subplot menjadi array 1 dimensi.

ax[0].imshow(citra1, cmap='gray')  # Menampilkan citra1 pada subplot indeks 0 dengan colormap 'gray'.
ax[0].set_title("Input Citra 1")  # Menetapkan judul subplot indeks 0 menjadi "Input Citra 1".

ax[1].imshow(citra2, cmap='gray')  # Menampilkan citra2 pada subplot indeks 1 dengan colormap 'gray'.
ax[1].set_title("Input Citra 1")  # Menetapkan judul subplot indeks 1 menjadi "Input Citra 1".

ax[2].imshow(output1, cmap='gray')  # Menampilkan output1 pada subplot indeks 2 dengan colormap 'gray'.
ax[2].set_title("Output Citra 1")  # Menetapkan judul subplot indeks 2 menjadi "Output Citra 1".

ax[3].imshow(output2, cmap='gray')  # Menampilkan output2 pada subplot indeks 3 dengan colormap 'gray'.
ax[3].set_title("Output Citra 2")  # Menetapkan judul subplot indeks 3 menjadi "Output Citra 2".
```
