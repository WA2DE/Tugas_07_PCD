# Praktikum 7 - Konvolusi

Import Library

```python
import matplotlib.pyplot as plt  # Mengimpor library matplotlib untuk visualisasi

 # Menggunakan inline backend untuk menampilkan plot di dalam notebook
%matplotlib inline

from skimage import data  # Mengimpor modul data dari library skimage
from skimage.io import imread  # Mengimpor fungsi imread dari modul io di library skimage
from skimage.color import rgb2gray  # Mengimpor fungsi rgb2gray dari modul color di library skimage
import numpy as np  # Mengimpor library numpy untuk manipulasi array dan matriks

import cv2  # Mengimpor library cv2 (OpenCV) untuk pengolahan citra
```

Load Image

```python
citra1 = imread(fname="mobil.tif")  # Membaca citra dengan menggunakan fungsi imread dan menyimpannya ke dalam variabel citra1
print(citra1.shape)  # Menampilkan dimensi citra menggunakan atribut shape

plt.imshow(citra1, cmap='gray')  # Menampilkan citra menggunakan fungsi imshow dari library matplotlib dengan colormap 'gray'
```

Proses Konvolusi

```python
kernel = np.array([[-1, 0, -1],  # Matriks kernel untuk operasi filter 2D
                   [0, 4, 0], 
                   [-1, 0, -1]])

citraOutput = cv2.filter2D(citra1, -1, kernel)  # Melakukan filter 2D pada citra menggunakan kernel yang telah ditentukan

fig, axes = plt.subplots(1, 2, figsize=(12, 12))  # Membuat subplot dengan 1 baris dan 2 kolom untuk menampilkan citra input dan output
ax = axes.ravel()  # Mengatur akses untuk subplot

ax[0].imshow(citra1, cmap='gray')  # Menampilkan citra input pada subplot pertama
ax[0].set_title("Citra Input")  # Mengatur judul subplot pertama
ax[1].imshow(citraOutput, cmap='gray')  # Menampilkan citra output pada subplot kedua
ax[1].set_title("Citra Output")  # Mengatur judul subplot kedua
```
