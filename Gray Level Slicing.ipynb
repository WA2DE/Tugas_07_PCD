# Praktikum 7.2 - Gray Level Slicing

Import Library

```python
import cv2  # Mengimpor library OpenCV untuk memanipulasi gambar.
import numpy as np  # Mengimpor library numpy untuk operasi array dan matriks.
from skimage import data  # Mengimpor submodul data dari library scikit-image.
import matplotlib.pyplot as plt  # Mengimpor submodul pyplot dari library matplotlib untuk membuat plot dan visualisasi data.
%matplotlib inline  # Menentukan bahwa plot akan ditampilkan secara langsung di notebook.
```

Gray Level Slice

```python
img = cv2.imread("gedung.tif", 0)  # Membaca gambar "gedung.tif" sebagai citra grayscale dan menyimpannya dalam variabel 'img'.

row, column = img.shape  # Mengambil dimensi baris dan kolom dari citra.

img1 = np.zeros((row,column),dtype = 'uint8')  # Membuat matriks nol dengan ukuran yang sama dengan citra menggunakan numpy dan menyimpannya dalam variabel 'img1'.

min_range = 10  # Menentukan nilai rentang minimum.
max_range = 60  # Menentukan nilai rentang maksimum.

for i in range(row):  # Melakukan perulangan untuk setiap baris citra.
    for j in range(column):  # Melakukan perulangan untuk setiap kolom citra.
        if img[i,j]>min_range and img[i,j]<max_range:  # Memeriksa apakah nilai piksel berada dalam rentang yang ditentukan.
            img1[i,j] = 255  # Jika iya, setel nilai piksel pada citra output menjadi 255 (putih).
        else:
            img1[i,j] = 0  # Jika tidak, setel nilai piksel pada citra output menjadi 0 (hitam).
```

Plot Image

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))  # Membuat subplot dengan ukuran 2x2 dan mengatur ukuran total figur.

ax = axes.ravel()  # Meratakan array subplot menjadi satu dimensi.

ax[0].imshow(img, cmap=plt.cm.gray)  # Menampilkan citra input dalam subplot indeks 0 dengan colormap gray.
ax[0].set_title("Citra Input")  # Menetapkan judul subplot indeks 0 sebagai "Citra Input".

ax[1].hist(img.ravel(), bins=256)  # Menampilkan histogram citra input dalam subplot indeks 1 dengan 256 bins.
ax[1].set_title('Histogram Input')  # Menetapkan judul subplot indeks 1 sebagai "Histogram Input".

ax[2].imshow(img1, cmap=plt.cm.gray)  # Menampilkan citra output dalam subplot indeks 2 dengan colormap gray.
ax[2].set_title("Citra Output")  # Menetapkan judul subplot indeks 2 sebagai "Citra Output".

ax[3].hist(img1.ravel(), bins=256)  # Menampilkan histogram citra output dalam subplot indeks 3 dengan 256 bins.
ax[3].set_title('Histogram Output')  # Menetapkan judul subplot indeks 3 sebagai "Histogram Output".
```
